```python
import requests
import pandas as pd
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
```

```python
data = resp.json()
```

```python
data[0].keys()
```

```python
issues = pd.DataFrame(data, columns=['number', 'title', 'labels', 'state'])
```
